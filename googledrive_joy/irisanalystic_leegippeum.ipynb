```
# data에서 같은 리스트로 묶인 [sepal length, sepal width, petal length, petal width]의 값들을 
# sepal length의 값들로만 이루어진 리스트와 같이 각각 나누어 리스트로 묶어줌.

sepal_length = data[:,0]
sepal_width = data[:,1]
petal_length = data[:,2]
petal_width = data[:,3]
```

```
# sepal length, sepal width, petal length, petal width의 각각의 mean, median, std, var, min, max를 구하기
```

```
# mean

mean_sepal_length = np.mean(sepal_length[:,], axis = 0)
mean_sepal_width = np.mean(sepal_width[:,], axis = 0)
mean_petal_length = np.mean(petal_length[:,], axis = 0)
mean_petal_width = np.mean(petal_width[:,], axis = 0)


# median

median_sepal_length = np.median(sepal_length[:,], axis = 0)
median_sepal_width = np.median(sepal_width[:,], axis = 0)
median_petal_length = np.median(petal_length[:,], axis = 0)
median_petal_width = np.median(petal_width[:,], axis = 0)


# std

std_sepal_length = np.std(sepal_length[:,], axis = 0)
std_sepal_width = np.std(sepal_width[:,], axis = 0)
std_petal_length = np.std(petal_length[:,], axis = 0)
std_petal_width = np.std(petal_width[:,], axis = 0)


# var

var_sepal_length = np.var(sepal_length[:,], axis = 0)
var_sepal_width = np.var(sepal_width[:,], axis = 0)
var_petal_length = np.var(petal_length[:,], axis = 0)
var_petal_width = np.var(petal_width[:,], axis = 0)


# min

min_sepal_length = np.min(sepal_length[:,], axis = 0)
min_sepal_width = np.min(sepal_width[:,], axis = 0)
min_petal_length = np.min(petal_length[:,], axis = 0)
min_petal_width = np.min(petal_width[:,], axis = 0)


# max

max_sepal_length = np.max(sepal_length[:,], axis = 0)
max_sepal_width = np.max(sepal_width[:,], axis = 0)
max_petal_length = np.max(petal_length[:,], axis = 0)
max_petal_width = np.max(petal_width[:,], axis = 0)
```

```
# mean, median, std, var, min, max 순서

len_sepal = mean_sepal_length, median_sepal_length, std_sepal_length, var_sepal_length, min_sepal_length, max_sepal_length
wid_sepal = mean_sepal_width, median_sepal_width, std_sepal_width, var_sepal_width, min_sepal_width, max_sepal_width
len_petal = mean_petal_length, median_petal_length, std_petal_length, var_petal_length, min_petal_length, max_petal_length
wid_sepal = mean_petal_width, median_petal_width, std_petal_width, var_petal_width, min_petal_width, max_petal_width


len_sepal, wid_sepal, len_petal, wid_sepal
```

```
plt.plot(x1,y1,'go--',label="Line Setosa")
plt.plot(x2,y2,'bo--',label="Line Versicolor")
plt.plot(x3,y3,'ro--',label="Line Virginica")
plt.title("Sepal")
plt.xlabel("length")
plt.ylabel("width")
plt.legend(['setosa','versicolor','virginica'])
plt.show()
```
